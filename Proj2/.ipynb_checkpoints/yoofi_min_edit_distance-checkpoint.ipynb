```python
w1 = 'intention'
w2 = 'execution'
if(w2[0]==w1[3]):
    print('hey')
```

```python
n = len(w1)
m = len(w2)
```

```python
matri = {}
for i in range(n+1):
    for j in range(m+1):
        matri[str(i)+":"+str(j)]=0
matri    
```

```python
matri['0:0']=0
for i in range(1,n+1):
    matri[str(i)+":0"] = matri[str(i-1)+":0"] + 1
for j in range(1,m+1):
    matri["0:"+str(j)] = matri["0:"+str(j-1)] + 1

# for i in range(1,10):
#     print(str(i)+":0 :"+ str(matri[str(i)+":0"]))
    
# for i in range(1,10):
#     print("0:"+str(i)+" :"+ str(matri["0:"+str(i)]))    
```

```python
for i in range(1,n+1):
    for j in range(1, m+1):
        if(w1[i-1]==w2[j-1]):
            sub = 0
        else:
            sub = 2
        matri[str(i)+":"+str(j)]=min((matri[str(i-1)+":"+str(j)])+1,(matri[str(i-1)+":"+str(j-1)])+sub,(matri[str(i)+":"+str(j-1)])+1)
```
